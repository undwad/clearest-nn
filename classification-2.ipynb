```octave
global ipynb = 'classification-2';

source('clearest-nn.m');
source('utils-logging.m');
source('utils-training.m');
source('utils-plotting.m');

function [X, Y] = generate_classN_samples(ranges, m)
    X = [];
    Y = [];
    n = rows(ranges);
    for i = 1:n
        rng = ranges(i, :);
        rng = repmat(rng, [m 1]);
        x   = randn(m, 2) + rng;
        y   = ones(m, 1) * i; 
        X   = [ X, x' ]; 
        Y   = [ Y, y' ]; 
    end
end

rand('state', 1);

classes = [1,2,3];

ranges   = [ -2 -2; 2 3; -1 1];
[X1, Y1] = generate_classN_samples(ranges, 30);
[X2, Y2] = generate_classN_samples(ranges, 30);
[X1, Y1] = shuffle(X1, Y1);
Y1oh     = onehots(Y1, classes);

printvar('classes');
printvar('ranges');

printsize('X1');
printsize('Y1');
printsize('X2');
printsize('Y2');

printsize('Y1oh');

BBB = minibatches(X1, Y1oh);

printsize('BBB');
printsize('BBB{1}.X');
printsize('BBB{1}.Y');
printsize('BBB{end}.X');
printsize('BBB{end}.Y');

plot_both('input', X1, Y1, X2, Y2, classes);
```
